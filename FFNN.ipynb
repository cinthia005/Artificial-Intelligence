**Objective:** The objective is to design, train, and evaluate a Feedforward Neural Network to classify handwritten digits from the MNIST dataset using Keras/TensorFlow in Google Colab.

**1. Load Libraries**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

**2. Load data**

```python
# Load MNIST dataset using Keras load_data() function
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Display the dataset information
print("\nDataset Information:")
print("Training images shape:", train_images.shape)
print("Training labels shape:", train_labels.shape)
print("Test images shape:", test_images.shape)
print("Test labels shape:", test_labels.shape)
print("Number of classes: 10 (digits 0-9)")
```

**3. Data Preprocessing**

```python
# Normalize pixel values to range [0, 1]
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

print("\nData normalized: Pixel values scaled to [0, 1]")

# Convert labels to categorical one-hot encoding
train_labels_cat = to_categorical(train_labels, 10)
test_labels_cat = to_categorical(test_labels, 10)

print("Labels converted to categorical format")

# Display sample images from the dataset
print("\nDisplaying sample images from MNIST dataset:")
plt.figure(figsize=(12, 4))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(train_images[i], cmap='gray')
    plt.title(f'Label: {train_labels[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

**4. Model Selection and Training**

```python
# Build Feedforward Neural Network Model
print("\nBuilding Feedforward Neural Network Model...")

model = Sequential([
    Flatten(input_shape=(28, 28)),      # Input layer: flatten 28x28 images to 784 pixels
    Dense(128, activation='relu'),      # First hidden layer: 128 neurons with ReLU
    Dense(64, activation='relu'),       # Second hidden layer: 64 neurons with ReLU
    Dense(10, activation='softmax')     # Output layer: 10 neurons with Softmax
])

print("Model architecture created successfully!")

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Model compiled with:")
print("- Optimizer: Adam")
print("- Loss function: Categorical Crossentropy")
print("- Metric: Accuracy")

# Display model architecture
print("\nModel Summary:")
model.summary()

# Train the model using fit function
print("\nStarting model training...")
history = model.fit(
    train_images,
    train_labels_cat,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)

print("Model training completed!")

# Plot training history
print("\nPlotting training history...")
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy During Training')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

**5. Model Testing**

```python
# Evaluate the model on test data using evaluate function
print("\nEvaluating model on test data...")
test_loss, test_accuracy = model.evaluate(test_images, test_labels_cat, verbose=0)

print("=== MODEL EVALUATION RESULTS ===")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Make 5 different predictions with test data
print("\n=== 5 PREDICTIONS ===")

for i in range(5):
    # Get prediction for a test sample
    prediction = model.predict(test_images[i:i+1], verbose=0)
    predicted_label = np.argmax(prediction)
    actual_label = test_labels[i]

    # Display prediction results
    print(f"\nPrediction {i+1}:")
    print(f"Model prediction: {predicted_label}")
    print(f"Actual value: {actual_label}")
    print(f"Confidence: {np.max(prediction):.4f}")

    # Show the corresponding image
    plt.figure(figsize=(3, 3))
    plt.imshow(test_images[i], cmap='gray')
    plt.title(f'Pred: {predicted_label}, Actual: {actual_label}')
    plt.axis('off')
    plt.show()

print("\n=== EVALUATION ===")

# Generate predictions for all test data
all_predictions = model.predict(test_images, verbose=0)
all_predicted_labels = np.argmax(all_predictions, axis=1)

# Classification report
print("\nClassification Report:")
print(classification_report(test_labels, all_predicted_labels))

# Confusion matrix
cm = confusion_matrix(test_labels, all_predicted_labels)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd')
plt.title('Confusion Matrix - MNIST Digit Classification')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Save the trained model
model.save('mnist_ffnn_model.h5')
print("\nModel saved as 'mnist_ffnn_model.h5'")

# Final summary
print("\n" + "="*50)
print("FINAL TRAINING SUMMARY")
print("="*50)
print("Network Architecture: 784-128-64-10")
print("Activation Functions:")
print("  - Hidden layers: ReLU")
print("  - Output layer: Softmax")
print("Loss Function: Categorical Crossentropy")
print("Optimizer: Adam")
print(f"Final Test Accuracy: {test_accuracy:.4f}")
print(f"Final Test Loss: {test_loss:.4f}")
print("Training completed successfully!")
print("="*50)
```
